```python
import numpy as np
import pandas as pd
```

```python
#atribui leitura a variavel df
df = pd.read_csv("groceries.csv", header= None)
#substitui o nan por 0 
df.replace(np.nan , 0, inplace = True)
df.head(10)
```

```python
#Função que retira os zeros da tabela pra facilitar
def removeAllZerosFromList(lista):
    return list(filter(lambda x: x != 0, lista))
```

```python
#transformando tudo em uma lista de transações, contendo todas as informações 
list_all_transaction = []
#index= o número de cada uma das linhas, row = o valor do conteudo de cada linha
for index, row in df.iterrows():
    #transformando cada linha do data frame em uma lista em vetor
    listOfTransaction = row.values.tolist()
    listOfTransaction = removeAllZerosFromList(listOfTransaction)

    list_all_transaction.append(listOfTransaction)

list_all_transaction[0]
```

```python
#importando a biblioteca mlxtend, para ajudar os dados antes de usar o Apriori
from mlxtend.preprocessing import TransactionEncoder
#transforma a lista de lista em um vetor array de true e false ou 1 e 0
te = TransactionEncoder()
te_ary = te.fit(list_all_transaction).transform(list_all_transaction)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
#chamamos o frequent_patterns e importamos o algoritmo apriori com as regras associativas
from mlxtend.frequent_patterns import apriori
#calcula quais são os itens mais frequentes que aparecem juntos
#df= DataFrame, min_support = quantos % da compras tem esse item, use_colnames = um parametro para mostrar o nome dos produtos
frequent_itemsets = apriori(df, min_support = 0.01, use_colnames = True)
#ordenar do maior suporte pro menor suporte , os itens que tem maior ocorrência
frequent_itemsets.sort_values(by=['support'],ascending = False)
```

```python
from mlxtend.frequent_patterns import association_rules
#verifica quem compra oque
#frequent_itemsets= os itens que são frequentemente comprados juntos, metric= a metrica vai ser a cofiança, min_threshold = %50 
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold= 0.4 )
#ordena pelo lift, quanto maior o lift mais forte é a associação de a e b.
rules.sort_values(by=['lift'], ascending = False).drop(['leverage', 'conviction'],axis=1)
```
